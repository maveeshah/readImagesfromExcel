```python
import zipfile
import cv2
import os
XLSname = "Book1.xlsx"

EmbeddedFiles = zipfile.ZipFile(XLSname).namelist()
ImageFiles = [F for F in EmbeddedFiles if F.count('.png') or F.count('.jpeg') or F.count('.jpg') ]
lim = len(ImageFiles)

new_images = []
for Image in ImageFiles:
    new = zipfile.ZipFile(XLSname).extract(Image)
    new_images.append(cv2.imread(new,0))

folder_path = r"D:\Office\Dataset Manipulation\Images"
for x in new_images:
    path, dirs, files = next(os.walk(folder_path))
    file_count = len(files)
    img_path= folder_path + '/' + str(file_count+1) + '.jpg'
    cv2.imwrite(img_path, x)
```
